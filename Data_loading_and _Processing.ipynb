```python
import pandas as pd

df=pd.read_csv(r"/Users/suriya/Ukesh_AIML_Projects/Amazon Music Clustering/Data/single_genre_artists.csv")
df.drop(['id_songs','name_song','name_artists'],axis=1,inplace=True)
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(15,13))

plt.show()
```

```python
feature_columns=['duration_ms', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness','liveness', 'valence', 'tempo']

featured_df=df[feature_columns]

featured_df
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaled_df=scaler.fit_transform(featured_df)

scaled_df=pd.DataFrame(scaled_df,columns=featured_df.columns)

scaled_df
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca=PCA()
pca.fit_transform(scaled_df)
np.cumsum(pca.explained_variance_ratio_)*100
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca=PCA(n_components=6)
x_pca=pca.fit_transform(scaled_df)
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans


model=KMeans()
visualizer=KElbowVisualizer(model,k=(2,10))
visualizer.fit(x_pca)
visualizer.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA


pca=PCA(n_components=6)
x_pca=pca.fit_transform(scaled_df)

for n in range(2,11):
    model=KMeans(n_clusters=n,random_state=42)
    model.fit_predict(x_pca)
    print(f"Num_Cluster:{n}, Silhoutte_score:{silhouette_score(x_pca,model.labels_)}")

```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

eps=[0.9,1,1.5,1.7]
for n in eps:
    model=DBSCAN(eps=n,min_samples=10)
    labels=model.fit_predict(x_pca)

    num_cluster=len(set(labels)) - (1 if -1 in labels else 0)

    if num_cluster>1:
        silhotte=silhouette_score(x_pca,labels)
        print(f"eps:{n}, clusters={num_cluster}, Silhoutte:{silhotte:.3f}")
    else:
        print(f"eps:{n}, clusters={num_cluster}")
```

```python
print(pd.Series(labels).value_counts().sort_index())
```

```python
pca=PCA(n_components=6)
x_pca=pca.fit_transform(scaled_df)

model=KMeans(n_clusters=3,random_state=42)
labels=model.fit_predict(x_pca)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plot_df=pd.DataFrame(x_pca[:, :2], columns=["PC1", "PC2"])
plot_df["Cluster"]=labels
plot_df.to_csv(r"/Users/suriya/Ukesh_AIML_Projects/Amazon Music Clustering/Data/PCA_Result.csv",index=False)

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=plot_df,
    x="PC1",
    y="PC2",
    hue="Cluster")

plt.show()
```

```python
cluster_profile=scaled_df.copy()
cluster_profile["Clusters"]=labels
cluster_profile=cluster_profile.groupby("Clusters").mean()
cluster_profile
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.heatmap(cluster_profile,annot=True,cmap='coolwarm',fmt='.2f')

plt.show()
```
